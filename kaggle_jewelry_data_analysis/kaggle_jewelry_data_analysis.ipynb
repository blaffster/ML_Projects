```python
# https://www.kaggle.com/datasets/mkechinov/ecommerce-purchase-history-from-jewelry-store

# Import data and rename columns

import pandas as pd

df = pd.read_csv('jewelry.csv')

df.rename( columns={"2018-12-01 11:40:29 UTC": "Date", "1924719191579951782": "Order_ID",
                   '1842195256808833386':'Product_ID', '1':'Quantity',
                  '1806829201890738522':'Category_ID', 'jewelry.earring':'Category',
                  '0':'BrandID', '561.51':'Price_USD', '1515915625207851155':'User_ID',
                  'Unnamed: 9':'Product_Gender', 'red':'Color',
                  'gold':'Metal', 'diamond':'Gem'}, inplace=True )
```

```python
# Remove unnecessary columns / unnecessary column info

df_new = df.drop(['Order_ID','Category_ID','User_ID'], axis=1)

df_new['Date'] = df_new['Date'].map(lambda x: str(x)[:-13])

df_new
```
