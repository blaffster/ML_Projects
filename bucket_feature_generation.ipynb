```python
# Import packages/data

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
from sklearn.metrics import mean_squared_error

train = pd.read_csv('train.csv')
valid = pd.read_csv('valid.csv')
test = pd.read_csv('test.csv')

train['date'] = pd.to_datetime(train['date'])
valid['date'] = pd.to_datetime(valid['date'])
test['date'] = pd.to_datetime(test['date'])

cols_to_keep = ['item_id','date','year','sell_price','units_sold']

train = train[cols_to_keep]
valid = valid[cols_to_keep]
test = test[cols_to_keep]

train
```

```python
# Function to handle leap years

def leap_year(df):
    
    ids = df.item_id.unique()

    df_single_item = df[df['item_id']==ids[0]]
    years = df_single_item.year.unique()

    if sum(years==2012) == 1:

        a = df_single_item[df_single_item['date']<'2012-02-29']

        b = df_single_item[ (df_single_item['date']>='2012-02-29') & (df_single_item['date']<'2016-02-29') ]
        b['date'] = b['date'] + pd.Timedelta('1 day')

        c = pd.concat([a,b])

    else:
        c = df_single_item[df_single_item['date']<'2016-02-29']

    if sum(years==2016) == 1:

        d = df_single_item[ df_single_item['date']>='2016-02-29' ]
        d['date'] = d['date'] + pd.Timedelta('1 day')

        e = pd.concat([c,d])

    else:

        d = df_single_item[df_single_item['date']>='2016-02-29']
        e = pd.concat([c,d])

    df_new = e.copy()

    for i in range(1,len(ids)):

        df_single_item = df[df['item_id']==ids[i]]
        years = df_single_item.year.unique()

        if sum(years==2012) == 1:

            a = df_single_item[df_single_item['date']<'2012-02-29']

            b = df_single_item[ (df_single_item['date']>='2012-02-29') & (df_single_item['date']<'2016-02-29') ]
            b['date'] = b['date'] + pd.Timedelta('1 day')

            c = pd.concat([a,b])

        else:
            c = df_single_item[df_single_item['date']<'2016-02-29']

        if sum(years==2016) == 1:

            d = df_single_item[ df_single_item['date']>='2016-02-29' ]
            d['date'] = d['date'] + pd.Timedelta('1 day')

            e = pd.concat([c,d])

        else:

            d = df_single_item[df_single_item['date']>='2016-02-29']
            e = pd.concat([c,d])

        df_new = pd.concat([df_new,e])
    
    return df_new
```

```python
train_new = leap_year(train)
valid_new = leap_year(valid)
test_new = leap_year(test)

train_new
```

```python
days_in_month = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]

train_new['day_of_yr'] = train_new['date'].apply( lambda x: x.day + sum(days_in_month[0:x.month-1]) )
valid_new['day_of_yr'] = valid_new['date'].apply( lambda x: x.day + sum(days_in_month[0:x.month-1]) )
test_new['day_of_yr'] = test_new['date'].apply( lambda x: x.day + sum(days_in_month[0:x.month-1]) )

train_new
```

```python
def single_year_bucket(df_single_item,year):
    
    df_single_item_single_year = df_single_item[df_single_item['year']==year]

    bucket_size = 5
    loops = int(365/bucket_size)

    k = 0
    start = k*bucket_size+1
    end = bucket_size

    df_slice = df_single_item_single_year[ (df_single_item_single_year['day_of_yr']>=start) & (df_single_item_single_year['day_of_yr']<=end) ]

    if df_slice.shape[0] > 0:

        df_slice['units_sold_'+str(bucket_size)] = df_slice.units_sold.sum()
        df_slice['sell_price_'+str(bucket_size)] = df_slice.sell_price.mean()
        df_slice = df_slice.drop(columns=['date','day_of_yr','sell_price','units_sold'])
        df_slice['bucket_num'] = k
        df_slice = df_slice.iloc[[0]]

    else:

        df_slice['units_sold_'+str(bucket_size)] = None
        df_slice['sell_price_'+str(bucket_size)] = None
        df_slice = df_slice.drop(columns=['date','day_of_yr','sell_price','units_sold'])

    for k in range(1,loops):

        start = k*bucket_size+1
        end = k*bucket_size+bucket_size

        df_dummy = df_single_item_single_year[ (df_single_item_single_year['day_of_yr']>=start) & (df_single_item_single_year['day_of_yr']<=end) ]

        if df_dummy.shape[0] > 0:

            df_dummy['units_sold_'+str(bucket_size)] = df_dummy.units_sold.sum()
            df_dummy['sell_price_'+str(bucket_size)] = df_dummy.sell_price.mean()
            df_dummy = df_dummy.drop(columns=['date','day_of_yr','sell_price','units_sold'])
            df_dummy['bucket_num'] = k
            df_dummy = df_dummy.iloc[[0]]

        else:

            df_dummy['units_sold_'+str(bucket_size)] = None
            df_dummy['sell_price_'+str(bucket_size)] = None
            df_dummy = df_dummy.drop(columns=['date','day_of_yr','sell_price','units_sold'])

        df_slice = pd.concat([df_slice,df_dummy])
    
    return df_slice
```
