```python
# Import packages/data

import pandas as pd
import time
import warnings
warnings.filterwarnings('ignore')

df = pd.read_csv('sales_train_evaluation.csv')
df
```

```python
# We're only going to look at store CA_1

df1 = df[df['store_id']=='CA_1']
df1 = df1.drop(columns=['store_id', 'state_id'])

df1
```

```python
# We can get rid of 'id' column

print( df1.id.nunique(), df1.item_id.nunique())
df2 = df1.drop(columns=['id'])

df2
```

```python
# There are no nulls

df2.isnull().sum().sum()
```

```python
# Calendar data cleaning
```

```python
################################################################################################################################
```

```python
# Import calendar csv, drop unnecessary columns, convert date format

cal_df = pd.read_csv('calendar.csv')
cal_df = cal_df.drop(columns=['snap_TX','snap_WI'])
cal_df['date'] = pd.to_datetime(cal_df['date'])

cal_df
```

```python
# Looks like we have some NaNs

cal_df.isnull().sum()
```

```python
# Replace NaNs with 'None'

cal_df = cal_df.where(pd.notnull(cal_df), 'None')
cal_df
```

```python
# Let's remove redundant column 'wday'

print(cal_df.weekday.nunique(), cal_df.wday.nunique())

cal_df = cal_df.drop(columns=['wday'])

cal_df
```

```python
# Add day column and transform/modify date info

cal_df['day'] = cal_df['date'].apply( lambda x: x.day )

cols = cal_df.columns.tolist()
cols = [cols[0]]+[cols[-1]]+cols[3:5]+cols[1:3]+cols[5:-1]
cal_df = cal_df[cols]

cal_df
```

```python
# Price data cleaning + EDA
```

```python
################################################################################################################################
```

```python
# Import price csv, only keep rows where store_id==CA_1

price_df = pd.read_csv('sell_prices.csv')
price_df = price_df[price_df['store_id']=='CA_1']
price_df = price_df.drop(columns=['store_id'])

price_df
```

```python
################################################################################################################################
```

```python
# Data transformation + merging
```

```python
################################################################################################################################
```

```python
# Let's recall what df2 looks like

df2
```

```python
# Determine rows/columns for data transformation

rows = df2.shape[0]
cols = df2.shape[1] - 4

rows, cols
```

```python
# Let's look at just the first row

a = df2.iloc[[0]]
a
```

```python
# Perform data transformation for only first row

start = time.time()

train_percent = 0.8
valid_percent = 0.1
test_percent = 0.1

for i in range(0,1):
    
    a = df2.iloc[[i]]
    values = a.to_numpy()[0][4:]
    
    b = a.loc[:, 'item_id':'d_1']
    b['d'] = 'd_1'
    b = b.rename(columns={"d_1": "units_sold"})
    
    for j in range(cols):
    
        c = a.loc[:, 'item_id':'cat_id']
        c['units_sold'] = values[j]
        c['d'] = 'd_'+str(j+2)

        b = pd.concat([b,c])
        

    df_new = b.merge(cal_df,on=['d'])
    df_new = df_new.merge(price_df,on=['item_id', 'wm_yr_wk'])
    
    rows_dummy = df_new.shape[0]

    stop1 = int(rows_dummy*train_percent)
    stop2 = stop1 + int(rows_dummy*valid_percent)

    train = df_new[0:stop1]
    valid = df_new[stop1:stop2]
    test = df_new[stop2:]

print(time.time()-start)
```

```python
print(train.shape[0] / df_new.shape[0], valid.shape[0] / df_new.shape[0], test.shape[0] / df_new.shape[0])
train
```

```python
# Perform data transformation for remaining rows, concatenate with first row only result

# for i in range(1,rows):
for i in range(1,100):
    
    a = df2.iloc[[i]]
    values = a.to_numpy()[0][4:]
    
    b = a.loc[:, 'item_id':'d_1']
    b['d'] = 'd_1'
    b = b.rename(columns={"d_1": "units_sold"})
    
    for j in range(cols):
    
        c = a.loc[:, 'item_id':'cat_id']
        c['units_sold'] = values[j]
        c['d'] = 'd_'+str(j+2)

        b = pd.concat([b,c])

    d = b.merge(cal_df,on=['d'])
    d = d.merge(price_df,on=['item_id', 'wm_yr_wk'])

    rows_dummy = d.shape[0]

    stop1 = int(rows_dummy*train_percent)
    stop2 = stop1 + int(rows_dummy*valid_percent)

    train_dummy = d[0:stop1]
    valid_dummy = d[stop1:stop2]
    test_dummy = d[stop2:]
    
    train = pd.concat([train,train_dummy])
    valid = pd.concat([valid,valid_dummy])
    test = pd.concat([test,test_dummy])
```

```python
print(train.shape[0] / (train.shape[0]+valid.shape[0]+test.shape[0]))
print(valid.shape[0] / (train.shape[0]+valid.shape[0]+test.shape[0]))
print(test.shape[0] / (train.shape[0]+valid.shape[0]+test.shape[0]))
```

```python
# Save train/validation/test data

train.to_csv('train.csv')
valid.to_csv('valid.csv')
test.to_csv('test.csv')
```
