```python
# https://www.kaggle.com/competitions/m5-forecasting-accuracy

# Import packages/data

import pandas as pd

df = pd.read_csv('sales_train_evaluation.csv')

df
```

```python
# We're only going to look at store CA_1

df1 = df[df['store_id']=='CA_1']
df1
```

```python
# We can get rid of 'id' column

len(df1.id.unique()), len(df1.item_id.unique())
```

```python
# Do we keep dept_id or just use cat_id?

df1.dept_id.unique(), df1.cat_id.unique()
```

```python
# Drop unnecessary columns

df2 = df1.drop(columns=['id', 'store_id', 'state_id'])
df2
```

```python
# There are no nulls

df2.isnull().sum().sum()
```

```python
################################################################################################################################
```

```python
# Import calendar csv, drop unnecessary columns, convert date format

cal_df = pd.read_csv('calendar.csv')
cal_df = cal_df.drop(columns=['snap_TX','snap_WI'])
cal_df['date'] = pd.to_datetime(cal_df['date'])

cal_df
```

```python
# Let's replace NaNs with 'None'

cal_df.isnull().sum()
```

```python
# Replace NaNs with 'None'

cal_df = cal_df.where(pd.notnull(cal_df), 'None')
cal_df
```

```python
# Let's remove redundant column 'wday'

cal_df.weekday.unique(), cal_df.wday.unique()
```

```python
# Remove redundant column 'wday'

cal_df = cal_df.drop(columns=['wday'])
cal_df
```

```python
# Transform/modify date info

cal_df['day'] = cal_df['date'].apply( lambda x: x.day )

cal_df = cal_df.drop(columns=['date'])

cols = cal_df.columns.tolist()
cols = [cols[4]] + [cols[-1]] + cols[2:4] + [cols[1]] + [cols[0]] + cols[5:10]
cal_df = cal_df[cols]

cal_df
```

```python
################################################################################################################################
```

```python
# Import price csv, only keep rows where store_id==CA_1

price_df = pd.read_csv('sell_prices.csv')
price_df = price_df[price_df['store_id']=='CA_1']
price_df = price_df.drop(columns=['store_id'])

price_df
```

```python
################################################################################################################################
```

```python
# Let's recall what df2 looks like

df2
```

```python
# Determine rows/columns for data transformation

rows = df2.shape[0]
cols = df2.shape[1] - 4

rows, cols
```

```python
# Perform data transformation for only first row

import warnings
warnings.filterwarnings('ignore')

for i in range(0,1):
    
    a = df2.iloc[[i]]
    values = a.to_numpy()[0][4:]
    
    b = a.loc[:, 'item_id':'d_1']
    b['d'] = 'd_1'
    b = b.rename(columns={"d_1": "units_sold"})
    
    for j in range(cols):
    
        c = a.loc[:, 'item_id':'cat_id']
        c['units_sold'] = values[j]
        c['d'] = 'd_'+str(j+2)

        b = pd.concat([b,c])
        

    df_new = b.merge(cal_df)
    df_new = df_new.merge(price_df,on=['item_id', 'wm_yr_wk'])

df_new
```

```python
# Perform data transformation for remaining rows, concatenate with first row only result

# for i in range(1,rows):
for i in range(1,50):
    
    a = df2.iloc[[i]]
    values = a.to_numpy()[0][4:]
    
    b = a.loc[:, 'item_id':'d_1']
    b['d'] = 'd_1'
    b = b.rename(columns={"d_1": "units_sold"})
    
    for j in range(cols):
    
        c = a.loc[:, 'item_id':'cat_id']
        c['units_sold'] = values[j]
        c['d'] = 'd_'+str(j+2)

        b = pd.concat([b,c])

    c = b.merge(cal_df)
    c = c.merge(price_df,on=['item_id', 'wm_yr_wk'])
    df_new = pd.concat([df_new,c])

df_new = df_new.drop(columns=['d'])
df_new
```

```python
################################################################################################################################
```

```python
dataframe = df_new.copy()
dataframe
```

```python
quantity_df = dataframe.copy()
quantity_df['quantity'] = 1
quantity_df = quantity_df.groupby(['item_id','month','year'])['quantity'].sum().reset_index()
quantity_df = quantity_df.rename(columns={'quantity':'days_in_month'})

quantity_df
```

```python
monthly_sales = dataframe.groupby(['item_id','month','year'])['units_sold'].sum().reset_index()
monthly_sales = monthly_sales.merge(quantity_df,on=['item_id','month','year'])

monthly_sales
```

```python
monthly_mean_sales = monthly_sales.copy()
monthly_mean_sales = monthly_mean_sales.rename(columns={'units_sold':'units_sold_monthly_mean'})
monthly_mean_sales['units_sold_monthly_mean'] = monthly_mean_sales['units_sold_monthly_mean'] / monthly_mean_sales['days_in_month']
monthly_mean_sales = monthly_mean_sales.drop(columns=['days_in_month'])

monthly_mean_sales
```

```python
overall_mean_monthly_sales = monthly_sales.groupby(['item_id','month'])['units_sold'].mean().reset_index()
overall_mean_monthly_sales = overall_mean_monthly_sales.rename(columns={"units_sold":"overall_units_sold_monthly_mean"})

overall_mean_monthly_sales
```

```python
dataframe_new = dataframe.merge(monthly_mean_sales,on=['item_id','month','year'])
dataframe_new = dataframe_new.merge(overall_mean_monthly_sales,on=['item_id','month'])
dataframe_new
```

```python
dataframe_new.to_csv('cleaned_data_50_prods_NEW.csv')
```
