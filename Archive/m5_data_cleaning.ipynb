```python
# Determine rows/columns for data transformation

rows = df2.shape[0]
cols = df2.shape[1] - 4

rows, cols
```

```python
# Let's look at just the first row

a = df2.iloc[[0]]
a
```

```python
# Perform data transformation for only first row

import time

import warnings
warnings.filterwarnings('ignore')

start = time.time()

for i in range(0,1):
    
    a = df2.iloc[[i]]
    values = a.to_numpy()[0][4:]
    
    b = a.loc[:, 'item_id':'d_1']
    b['d'] = 'd_1'
    b = b.rename(columns={"d_1": "units_sold"})
    
    for j in range(cols):
    
        c = a.loc[:, 'item_id':'cat_id']
        c['units_sold'] = values[j]
        c['d'] = 'd_'+str(j+2)

        b = pd.concat([b,c])
        

    df_new = b.merge(cal_df,on=['d'])
    df_new = df_new.merge(price_df,on=['item_id', 'wm_yr_wk'])

print(start-time.time())

df_new
```

```python
# Perform data transformation for remaining rows, concatenate with first row only result

# for i in range(1,rows):
for i in range(1,250):
    
    a = df2.iloc[[i]]
    values = a.to_numpy()[0][4:]
    
    b = a.loc[:, 'item_id':'d_1']
    b['d'] = 'd_1'
    b = b.rename(columns={"d_1": "units_sold"})
    
    for j in range(cols):
    
        c = a.loc[:, 'item_id':'cat_id']
        c['units_sold'] = values[j]
        c['d'] = 'd_'+str(j+2)

        b = pd.concat([b,c])

    d = b.merge(cal_df,on=['d'])
    d = d.merge(price_df,on=['item_id', 'wm_yr_wk'])
    df_new = pd.concat([df_new,d])

df_new = df_new.drop(columns=['d'])
df_new
```

```python
# Import cleaned data

df = pd.read_csv('cleaned_data_100_prods.csv')
df
```

```python
# Create train/test split

df2 = df.copy()

train_df = df2[df2['date'] < '2015-07-01']
test_df = df2[df2['date'] >= '2015-07-01']

train_df.shape[0] / df2.shape[0], test_df.shape[0] / df2.shape[0]
```

```python
train_df
```

```python
# Determine units sold for a given item, in a given month, of a given year

monthly_sales = train_df.groupby(['item_id','month','year'])['units_sold'].sum().reset_index()
monthly_sales
```

```python
# Determine mean units sold for a given item, in a given month, by averaging sales of product in given month over all years

overall_mean_monthly_sales = monthly_sales.groupby(['item_id','month'])['units_sold'].mean().reset_index()
overall_mean_monthly_sales = overall_mean_monthly_sales.rename(columns={"units_sold":"overall_units_sold_monthly_mean"})

overall_mean_monthly_sales
```

```python
# Add new features to test/train dfs

train_new = train_df.merge(overall_mean_monthly_sales,on=['item_id','month'])
test_new = test_df.merge(overall_mean_monthly_sales,on=['item_id','month'])

train_new
```

```python
# Feature engineering --> unit_sold_monthly_sum_rolling_mean_3
```

```python
################################################################################################################################
```

```python
import pandas as pd

df = pd.read_csv('cleaned_data_50_prods_NEW.csv')
df
```

```python
# Determine units sold for a given item, in a given month, of a given year

monthly_sales = df.groupby(['item_id','month','year'])['units_sold'].sum().reset_index()
monthly_sales
```

```python
a = monthly_sales.copy()
rows = a.shape[0]
dates = []

for i in range(rows):
    y = a.iloc[i]['month']
    z = a.iloc[i]['year']
    out = str(z)+'-'+str(y)+'-1'
    dates.append(out)
```

```python
a['date'] = dates
a['date'] = pd.to_datetime(a['date'])
a = a.sort_values(by=['date'])
# a = a.drop(columns=['date'])

a
```

```python
item_ids = a.item_id.unique()

for i in range(0,1):
    b = a[a['item_id']==item_ids[i]]
    b['unit_sold_monthly_sum_rolling_mean_3'] = b['units_sold'].rolling(3).mean()

b
```

```python
import warnings
warnings.filterwarnings('ignore')

rows = a.item_id.nunique()

for i in range(1,rows):
    c = a[a['item_id']==item_ids[i]]
    c['unit_sold_monthly_sum_rolling_mean_3'] = c['units_sold'].rolling(3).mean()
    b = pd.concat([b,c])

b
```

```python
b = b.where(pd.notnull(b), 0)
b = b.drop(columns=['units_sold','date'])

b
```

```python
df2 = df.merge(b,on=['item_id','month','year'])
df2
```
